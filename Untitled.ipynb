```python
class Company:
    aprice=0.0
    acount=0.0
    acount_true=0
    periods={} #3->97 24->18
```

```python
companies = {}
files = glob.glob('./pdfs/out/*.out')
for file in files:
    with open(file) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=' ')
        for row in csv_reader:
            companies[row[1]] = Company()
            companies[row[1]].aprice=row[2]
files = glob.glob('./pdfs/out/*.out2')
for file in files:
    with open(file) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=' ')
        for row in csv_reader:
            try:
                companies[row[1]].acount=row[2]
                companies[row[1]].acount_true=row[2]*1000000
            except:
                pass
            
            

```
